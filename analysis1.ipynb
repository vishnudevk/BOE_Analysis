Gross Written Premium(GWP) is the income to the firm. <br>
Bigger the GWP bigger the firm is. <br>

```python
gwp_2020 = df_sub[["GWP-2020YE"]]
gwp_2020.index = df_sub["Firm"]


print("Finding invalid outliers for GWP-2020YE")
print("Max ",gwp_2020["GWP-2020YE"].max())
print("Min ",gwp_2020["GWP-2020YE"].min()) #There are some zeors and few negatives in the data. Is it valid?
print("Median ",gwp_2020["GWP-2020YE"].median())
print("Mean ",gwp_2020["GWP-2020YE"].mean())

sns.histplot(gwp_2020, bins = 10).set(title='GWP distribution') #Histogram for visualising the data distribution 
plt.show()

#plt.figure(figsize=(20,5))
plt.xlim(-200, 500)
sns.boxplot(gwp_2020["GWP-2020YE"]).set(title='GWP distribution')
#The box plot is skewed and outliers are marked
plt.show()


Q1,Q3 = np.percentile(gwp_2020 , [25,75])
IQR = Q3 - Q1
ul = Q3+1.5*IQR
ll = Q1-1.5*IQR
outliers = (gwp_2020 > ul) | (gwp_2020 < ll)

outlier_vals =  gwp_2020[outliers].dropna().sort_values("GWP-2020YE",ascending=False) #List of of outliers ordered by GWP value
outlier_vals.head(10)
```

```python
df_sub_tmp = df_sub 
```

```python
df_sub = df_sub_tmp
df_sub = df_sub.drop(['Firm'], axis=1)

#df_sub = df_sub[[c for c in df_sub if c.startswith('NCR')]]
df_sub = df_sub[(df_sub > 0).all(1)] #Droping all the zero and negative rows since we are not sure if they are valid.
#having 1/3 elements being replaced by median wont be good for ML model

df_sub=(df_sub-df_sub.min())/(df_sub.max()-df_sub.min()) #Normalize the data in 0-1 range


from sklearn.model_selection import train_test_split
train, test = train_test_split(df_sub, test_size=0.2 , random_state=50)


train_labels = train['NCR-2020YE']
test_lables = test['NCR-2020YE']

train_features = train[[c for c in train if  not c.endswith('2020YE')]]
test_features = test[[c for c in test if  not c.endswith('2020YE')]]
```

```python
import tensorflow as tf
from os.path import exists

model = tf.keras.models.Sequential([
  tf.keras.layers.Dense(20 , activation='relu'),
  tf.keras.layers.Dropout(0.2),
  #tf.keras.layers.Dense(50, activation='relu'),
  #tf.keras.layers.Dropout(0.2),
  #tf.keras.layers.Dense(10, activation='relu'),
  #tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(1, activation='relu')
])


model.compile(optimizer='adam',
              loss='mse')


if exists("saved_weights.index"):
  model.load_weights("saved_weights")
else:
  model.fit(train_features, train_labels,  validation_data=(test_features, test_lables),epochs=12)
```

```python
predictions = np.squeeze(model.predict(test_features))
sns.lineplot(y=test_lables,x=range(len(predictions)),label="Actal NCR-2020")
sns.lineplot(y=predictions,x=range(len(predictions)),label="Predicted NCR-2020")
```

Doing PCA to reduce dimensions.
Doing K means clustering to find outliers       


```python
from sklearn.preprocessing import StandardScaler
#scaled_features = StandardScaler().fit_transform(df.values)

df_sub = df_sub_tmp
firms = df_sub['Firm']
df_sub = df_sub.drop(['Firm'], axis=1)
df_sub.index = firms

#
scaled_features = StandardScaler().fit_transform(df_sub.values)
df_sub = pd.DataFrame(scaled_features, index=df_sub.index, columns=df_sub.columns)
df_sub
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_sub)

pca_df = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
pca_df.index = firms


sns.scatterplot(x=pca_df['principal component 1'], y = pca_df['principal component 2'])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=0).fit(pca_df)

clusters = kmeans.predict(pca_df)
clusters[clusters > 0] = 1 #combining cluster 1 and 2 as outlier

pca_df["outlier"] = clusters


sns.scatterplot(x='principal component 1', y = 'principal component 2', data=pca_df, hue="outlier", palette=['green','orange']).set(title='Outliers from PCA+ K means analysis') 
```
